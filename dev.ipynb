```python
import atomview as aw
from atomview.core import view
from atomview.utils import _load_structure
```

```python
def fetch_structure(pdb_id):
    import requests
    import os
    import biotite.structure as struc
    import biotite.structure.io as strucio
    import biotite.structure.io.pdb as pdb

    url = f"https://files.rcsb.org/download/{pdb_id}.cif"
    response = requests.get(url)
    pdb_path = f"{pdb_id}.cif"
    with open(pdb_path, "wb") as f:
        f.write(response.content)



aa = _load_structure('tests/data/4hhb.cif')
aa.get_annotation_categories()
```
